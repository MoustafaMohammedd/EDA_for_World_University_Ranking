```python
mm=df[["Institution","Score","Year"]].set_index("Year").sort_values(by="Score",ascending=False).nlargest(15,"Score")
mm
```

```python
mm["Institution"].unique()
```

```python
institutions =mm["Institution"].unique()

for institution in institutions:
    plt.plot(df["Year"].unique(), df[df["Institution"] == institution]["Score"], label=institution)
plt.title("the best")
plt.xlabel("the Year")
plt.ylabel("the score")
    
plt.legend()
plt.show()
```

```python
institutions =mm["Institution"].unique()

for institution in institutions:
    plt.scatter(df["Year"].unique(), df[df["Institution"] == institution]["Score"], label=institution)
plt.title("the best")
plt.xlabel("the Year")
plt.ylabel("the score")
    
plt.legend()
plt.show()
```
